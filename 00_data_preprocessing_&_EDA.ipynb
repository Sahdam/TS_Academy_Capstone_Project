```python
import numpy as np

fig, axes = plt.subplots(2, 1, figsize=(14, 10))

shock_date = pd.Timestamp('2023-06-01')

for ax, col, title, color in zip(
    axes,
    ['basket_cost', 'food_cpi'],
    ['Basket Cost CUSUM', 'CPI Food CUSUM'],
    ['#1B3A6B', '#0D6E8C']
):
    series = data_final[col].dropna()


    mean   = series.mean()
    cusum  = (series - mean).cumsum()


    std    = cusum.std()
    upper  =  2 * std
    lower  = -2 * std


    ax.plot(cusum.index, cusum.values, color=color, linewidth=2, label='CUSUM')
    ax.axhline(upper, color='orange', linestyle='--', linewidth=1.2, label=f'+2 bound ({upper:,.0f})')
    ax.axhline(lower, color='orange', linestyle='--', linewidth=1.2, label=f'-2 bound ({lower:,.0f})')
    ax.axhline(0,     color='grey',   linestyle='-',  linewidth=0.8, alpha=0.5)


    ax.axvspan(shock_date, series.index[-1], alpha=0.08, color='red')
    ax.axvline(shock_date, color='red', linestyle='--', linewidth=1.8, label='Fuel Subsidy Removal (Jun 2023)')


    peak_idx = cusum.abs().idxmax()
    peak_val = cusum[peak_idx]
    ax.scatter(peak_idx, peak_val, color='red', zorder=5, s=80)
    ax.annotate(f'Max deviation\n{peak_idx.strftime("%b %Y")}',
                xy=(peak_idx, peak_val),
                xytext=(peak_idx - pd.DateOffset(months=18), peak_val * 0.75),
                arrowprops=dict(arrowstyle='->', color='red'),
                fontsize=9, color='red')

    ax.set_title(title, fontsize=13, fontweight='bold')
    ax.set_ylabel('Cumulative Sum of Deviations')
    ax.legend(loc='upper left', fontsize=9)
    ax.grid(alpha=0.3)

plt.suptitle('CUSUM Analysis  Detecting Structural Break', fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('cusum_plot.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
#  Weighted contributions (mirrors basket_price formula) 
weights = {
    'eggs': 1,
    'brown_beans': 2,
    'bread': 8,
    'chicken': 2,
    'garri': 3,
    'maize': 1,
    'onions_fresh': 1,
    'palm_oil': 3,
    'local_rice': 5,
    'foriegn_rice': 5,
    'tomatoes': 3,
    'vegetable_oil': 5,
    'yam': 3,
    'potatoes': 2
}

contrib = pd.DataFrame(index=data_final.index)
for col, w in weights.items():
    contrib[col] = data_final[col] * w

contrib = contrib.dropna()

labels = [c.replace('_', ' ').title() for c in weights.keys()]

fig, ax = plt.subplots(figsize=(16, 8))

ax.stackplot(
    contrib.index,
    [contrib[c] for c in weights.keys()],
    labels=labels,
    colors=colors,
    alpha=0.85
)

ax.axvline(shock_date, color='white', linestyle='--',
           linewidth=2, label='Subsidy Removal (Jun 2023)')
ax.annotate('Fuel Subsidy Removal\nJun 2023',
            xy=(shock_date, contrib.sum(axis=1).max() * 0.75),
            xytext=(pd.Timestamp('2021-01-01'), contrib.sum(axis=1).max() * 0.80),
            arrowprops=dict(arrowstyle='->', color='white', lw=1.5),
            fontsize=9, color='white', fontweight='bold')

ax.set_title('Commodity Contributions to Basket Price (Weighted)',
             fontsize=14, fontweight='bold', pad=15)
ax.set_ylabel('Weighted Price Contribution ()', fontsize=11)
ax.set_xlabel('')
ax.grid(alpha=0.2, linestyle='--', axis='y')


ax.legend(loc='upper left', bbox_to_anchor=(1.01, 1),
          fontsize=9, frameon=True, title='Commodity', title_fontsize=10)

plt.tight_layout()
plt.savefig('commodity_stacked.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
pre_shock  = contrib[contrib.index <  shock_date].mean()
post_shock = contrib[contrib.index >= shock_date].mean()
change_pct = ((post_shock - pre_shock) / pre_shock * 100).sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(10, 7))

bars = ax.barh(
    [l.replace('_', ' ').title() for l in change_pct.index],
    change_pct.values,
    color=['#166534' if v < 0 else '#991B1B' for v in change_pct.values],
    edgecolor='white', linewidth=0.5
)

# Value labels
for bar, val in zip(bars, change_pct.values):
    ax.text(val + 1, bar.get_y() + bar.get_height() / 2,
            f'{val:.1f}%', va='center', fontsize=9,
            color='#1B3A6B', fontweight='bold')

ax.axvline(0, color='grey', linewidth=0.8)
ax.set_title('% Change in Weighted Contribution\nPre vs Post Fuel Subsidy Removal (Jun 2023)',
             fontsize=13, fontweight='bold')
ax.set_xlabel('% Change in Average Weighted Contribution', fontsize=10)
ax.grid(alpha=0.3, axis='x', linestyle='--')

plt.tight_layout()
plt.savefig('commodity_change.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
from scipy import stats

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

shock_date = pd.Timestamp('2023-06-01')

colors_scatter = data_final.index.map(
    lambda d: '#991B1B' if d >= shock_date else '#1B3A6B'
)

for ax, xvar, xlabel in zip(
    axes,
    ['exchange_rate', 'fuel_price'],
    ['Exchange Rate (/USD)', 'Fuel Price (/litre)']
):
    data = data_final[['basket_cost', xvar]].dropna()
    period_colors = data.index.map(
        lambda d: '#991B1B' if d >= shock_date else '#1B3A6B'
    )


    ax.scatter(data[xvar], data['basket_cost'],
               c=period_colors, alpha=0.75, s=55, edgecolors='white', linewidth=0.4)


    slope, intercept, r, p, _ = stats.linregress(data[xvar], data['basket_cost'])
    x_line = np.linspace(data[xvar].min(), data[xvar].max(), 200)
    ax.plot(x_line, slope * x_line + intercept,
            color='black', linewidth=1.8, linestyle='--',
            label=f'Overall fit  r={r:.2f}  p={p:.3f}')


    pre  = data[data.index <  shock_date]
    post = data[data.index >= shock_date]

    if len(pre) > 2:
        s1, i1, r1, p1, _ = stats.linregress(pre[xvar], pre['basket_cost'])
        x1 = np.linspace(pre[xvar].min(), pre[xvar].max(), 100)
        ax.plot(x1, s1*x1 + i1, color='#1B3A6B', linewidth=1.5,
                label=f'Pre-shock  r={r1:.2f}')


    if len(post) > 2:
        s2, i2, r2, p2, _ = stats.linregress(post[xvar], post['basket_cost'])
        x2 = np.linspace(post[xvar].min(), post[xvar].max(), 100)
        ax.plot(x2, s2*x2 + i2, color='#991B1B', linewidth=1.5,
                label=f'Post-shock  r={r2:.2f}')


    import matplotlib.patches as mpatches
    pre_patch  = mpatches.Patch(color='#1B3A6B', label='Pre-shock (before Jun 2023)')
    post_patch = mpatches.Patch(color='#991B1B', label='Post-shock (after Jun 2023)')
    handles, lbls = ax.get_legend_handles_labels()
    ax.legend(handles=handles + [pre_patch, post_patch], fontsize=8, loc='upper left')

    ax.set_xlabel(xlabel, fontsize=11)
    ax.set_ylabel('Basket Cost ()', fontsize=11)
    ax.set_title(f'Basket Cost vs {xlabel}', fontsize=12, fontweight='bold')
    ax.grid(alpha=0.3, linestyle='--')

plt.suptitle('Basket Cost Relationships with Macro Drivers',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('scatter_plots.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 10))

shock_date = pd.Timestamp('2023-06-01')

for ax, xvar, color, title in zip(
    axes,
    ['exchange_rate', 'fuel_price'],
    ['#0D6E8C', '#B45309'],
    ['Rolling 12-Month Correlation: Basket Cost vs Exchange Rate',
     'Rolling 12-Month Correlation: Basket Cost vs Fuel Price']
):
    rolling_corr = data_final['basket_cost'].rolling(window=12).corr(data_final[xvar])

    ax.plot(rolling_corr.index, rolling_corr.values,
            color=color, linewidth=2.2, label='12-month rolling correlation')


    ax.fill_between(rolling_corr.index, rolling_corr.values,
                    alpha=0.15, color=color)


    ax.axhline(1.0,  color='green', linestyle=':', linewidth=1, alpha=0.6, label='Perfect positive (1.0)')
    ax.axhline(0.0,  color='grey',  linestyle='-', linewidth=0.8, alpha=0.5, label='No correlation (0.0)')
    ax.axhline(0.7,  color='green', linestyle='--', linewidth=1, alpha=0.4, label='Strong threshold (0.7)')


    ax.axvline(shock_date, color='red', linestyle='--',
               linewidth=1.8, label='Subsidy Removal (Jun 2023)')
    ax.axvspan(shock_date, rolling_corr.index[-1],
               alpha=0.07, color='red')


    pre_avg  = rolling_corr[rolling_corr.index <  shock_date].mean()
    post_avg = rolling_corr[rolling_corr.index >= shock_date].mean()

    ax.annotate(f'Pre-shock avg\nr = {pre_avg:.2f}',
                xy=(pd.Timestamp('2020-01-01'), pre_avg),
                fontsize=9, color='#1B3A6B', fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.7))

    ax.annotate(f'Post-shock avg\nr = {post_avg:.2f}',
                xy=(pd.Timestamp('2024-01-01'), post_avg),
                fontsize=9, color='#991B1B', fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.7))

    ax.set_ylim(-0.2, 1.1)
    ax.set_ylabel('Pearson Correlation (r)', fontsize=11)
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.legend(loc='lower left', fontsize=8)
    ax.grid(alpha=0.3, linestyle='--')

plt.suptitle('Did the Macro-Basket Relationship Strengthen After the 2023 Shock?',
             fontsize=13, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('rolling_correlation.png', dpi=150, bbox_inches='tight')
plt.show()
```
