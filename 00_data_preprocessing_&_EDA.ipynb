```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/Sahdam/TS_Academy_Capstone_Project/refs/heads/main/Inflation_data.csv"
df_inf = pd.read_csv(url, parse_dates=["Date"], index_col="Date")
df_inf.head()
```

```python
url = "https://raw.githubusercontent.com/Sahdam/TS_Academy_Capstone_Project/refs/heads/main/Selected_Food_Price.csv"

df_selected = pd.read_csv(url, parse_dates=["Date"], index_col="Date")
df_selected.head()
```

```python
url = "https://raw.githubusercontent.com/Sahdam/TS_Academy_Capstone_Project/refs/heads/main/PMS%20Price.csv"
df_pms = pd.read_csv(url, parse_dates=["Date"], index_col="Date")
df_pms.head()
```

```python
url = "https://raw.githubusercontent.com/Sahdam/TS_Academy_Capstone_Project/refs/heads/main/Exchange%20rate.csv"
df_exchange = pd.read_csv(url, parse_dates=["Date"], index_col="Date")
df_exchange.head()
```

```python
df_exchange.tail()
```

```python
data_1 = pd.merge(df_inf, df_pms, on="Date")
data_1.head()
```

```python
data_2 = pd.merge(data_1, df_exchange, on="Date")
data_2.head()
```

```python
data_final = pd.merge(data_2, df_selected, on="Date")
data_final.head()
```

```python
data_final.info()
```

```python
data_final.drop(columns=['All Items', 'Food'], inplace=True)
```

```python
data_final.info()
```

```python
col_obj = data_final.select_dtypes("object").columns.to_list()
col_obj
```

```python
for i in col_obj:
  data_final[i] = data_final[i].str.replace(",","").astype(float)
```

```python
data_final.info()
```

```python
data_final.isnull().sum()
```

```python
inter_col = (data_final.isnull().sum()>0).sort_values(ascending=False).head(14).index.to_list()
inter_col
```

```python
for i in inter_col:
  data_final[i] = data_final[i].interpolate(method="time")
```

```python
data_final.isna().sum()
```

```python
data_final.rename(columns={
    "All_item_adjusted": "all_cpi",
    "Food_adjusted": "food_cpi",
    "PMS Price": "fuel_price",
    "ifemDollar": "exchange_rate",
    "Agric hen eggs, (a Crate of 30 pieces)":"eggs",
    "Beans Brown": "brown_beans",
    "Bread (sliced), 450g":"bread",
    "Chicken meat (Frozen)":"chicken",
    "Garri white":"garri",
    "Maize (Corn) Grains (White) sold loose":"maize",
    "Onions, fresh": "onion",
    "Palm oil, 75cl":"palm_oil",
    "Rice Local, short-Grained":"local_rice",
    "Rice Long-Grained (Imported)":"foriegn_rice",
    "Tomatoes, fresh":"tomatoes",
    "Vegetable Oil, 75cl":"vegetable_oil",
    "Yam Tuber":"yam",
    "Sweet potatoes":"potatoes"
}, inplace=True)
```

```python
data_final.head()
```

```python
data_final.describe()
```

```python
data_final.index
```

```python
data_final = data_final.asfreq(freq="ME")
```

```python
data_final.index
```

```python
data_final["food_cpi_mom"] = data_final["food_cpi"].pct_change()*100
data_final.head()
```

```python
data_final.columns = (
    data_final.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace(",", "")
    .str.replace("(", "")
    .str.replace(")", "")
)
```

```python
data_final['food_cpi_yoy'] = data_final['food_cpi'].pct_change(12)*100
data_final.head()
```

```python
data_final.columns = (
    data_final.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace(",", "")
    .str.replace("(", "")
    .str.replace(")", "")
)
```

```python
data_final.columns
```

```python
data_final["basket_cost"] = data_final['eggs'] + 2 * data_final['brown_beans']+ 8 * data_final['bread'] + 2 * data_final['chicken'] + 3 * data_final['garri']+ data_final['maize'] + data_final['onions_fresh']+ 3 * data_final['palm_oil']+ 5 * data_final['local_rice'] + 5 * data_final['foriegn_rice'] + 3 * data_final['tomatoes'] + 5 * data_final['vegetable_oil'] + 3* data_final['yam'] + 2 * data_final['potatoes']
```

```python
data_final.head()
```

```python
data_final.tail()
```

```python
data_final["basket_cost_mom"] = data_final["basket_cost"].pct_change()*100
data_final.head()
```

```python
data_final["basket_cost_yoy"] = data_final["basket_cost"].pct_change(12)*100
data_final.head()
```

```python
plt.figure(figsize=(13, 5))
plt.plot(data_final["food_cpi"], label="food_cpi", color="red", linewidth=2)
plt.xlabel("Date")
plt.ylabel("CPI")
plt.title("Food CPI [Inflation]")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(13, 5))
plt.plot(data_final["food_cpi"], label="food_cpi", color="red", linewidth=2)
plt.plot(data_final["fuel_price"], label="fuel price", color="blue", linewidth=2)
plt.plot(data_final["exchange_rate"], label="exchange rate", color="green", linewidth=2)
plt.xlabel("Date")
plt.ylabel("CPI")
plt.title("Food CPI [Inflation]|Fuel Price|Exchange Rate")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(13, 5))
plt.plot(data_final["food_cpi_mom"], label="food_cpi_mom", color="red", linewidth=2)
plt.plot(data_final["food_cpi_yoy"], label="food_cpi_yoy", color="blue", linewidth=2)
plt.xlabel("Date")
plt.ylabel("CPI mom & yoy")
plt.title("Food CPI [Inflation] Monthly and Yearly")
plt.legend(loc="upper left")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
